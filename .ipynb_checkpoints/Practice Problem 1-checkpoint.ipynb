```python
from sklearn.datasets import load_iris
```

```python
data3=load_iris()
```

```python
# Convert the dataset to a pandas DataFrame
iris = pd.DataFrame(data=data3.data, columns=data3.feature_names)

# Add a column for the target variable
iris['target'] = data3.target

# Print the first few rows of the DataFrame
print(iris.head())
```

```python
import matplotlib.pyplot as plt
#b
# assuming df is your DataFrame
for feature in iris.columns:
    if feature == 'target':  # target variable
        plt.hist(iris[feature], bins=10)
        plt.title(feature)
        plt.show()
    else:  # feature variables
        plt.hist(iris[feature], bins=10)
        plt.title(feature)
        plt.show()
```

```python
#c
corrmat = iris.corr(method = "spearman")
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(corrmat,annot=True)
```

```python
#e
from sklearn import preprocessing 
# print the normalized data
norm1= preprocessing.normalize(data3.data)
norm1
```

```python
#d
iris_df = pd.DataFrame(data=data3.data, columns=data3.feature_names)
iris_df['target'] = iris.target

# Generate pairplot
sns.pairplot(iris_df, hue='target', palette='Set2')
plt.show()
```
